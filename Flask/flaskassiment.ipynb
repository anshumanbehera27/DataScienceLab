#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

##### what is falsk framework?

Flask is a micro web framework written in Python. It is classified as a microframework because it does not require particular tools or libraries. It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions.

Easy to get started: Flask is easy to learn and has a simple and intuitive syntax, making it easy to get started with web development.

Lightweight: Flask is a lightweight framework and does not require much in terms of dependencies or overhead, which makes it a good choice for small to medium-sized projects.

Flexible: Flask is a flexible framework that can be customized and extended in many ways to suit the needs of a particular project.

Modular design: Flask's modular design allows developers to use only the features they need, without having to worry about unnecessary complexity.

Built-in development server: Flask comes with a built-in development server that makes it easy to test and debug applications during the development process.

Integration with other libraries: Flask can be easily integrated with other libraries and tools, such as SQLAlchemy, WTForms, and Jinja2, to add additional functionality to an application.

### What is App routing in Flask?

In Flask, app routing refers to the process of mapping URLs to the functions that handle them. It allows you to define different routes for your web application and associate them with specific functions, so that when a user requests a particular URL, the corresponding function is called and returns a response

#### what is use of app.route()?
The app.route() decorator in Flask is used to define the URL route for a particular view function. When a user makes a request to a URL that matches the route, Flask will execute the associated view function and return its response to the user
Flask supports a wide range of URL patterns, including static routes, dynamic routes with parameters, and even regular expression routes. The routing system also supports HTTP methods, so you can define different routes for different HTTP methods like GET, POST, PUT, DELETE, etc

###  Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building. This function is used to generate a URL for a specific function based on the function name.

The url_for() function takes the name of the endpoint function as its first argument, and any number of keyword arguments, each representing a variable part of the URL. For example, if you have a function named home() that handles the root URL of your application, you can generate the URL for this function using url_for('home')
